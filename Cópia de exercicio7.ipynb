## 4. KPIs globais da operação

Com base nas “4 Métricas de Ouro do Varejo” apresentadas na aula, calcule os seguintes KPIs globais da base inteira: fileciteturn5file0

- **Receita Total**
- **Lucro Total**
- **Margem de Lucro (%)**
- **Ticket Médio**

### Orientação
- Pense primeiro na fórmula de cada KPI
- Escreva os cálculos em células separadas ou em uma pequena tabela-resumo
- Se quiser, formate os resultados para facilitar a leitura


```python
# Receita Total
receita_total = df['valor_venda'].sum()

# Lucro Total
lucro_total = df['lucro'].sum()

# Margem de Lucro (%)
margem_lucro = (lucro_total / receita_total) * 100

# Ticket Médio
ticket_medio = receita_total / len(df)

# Exibir resultados formatados
print(f"Receita Total: R$ {receita_total:,.2f}")
print(f"Lucro Total: R$ {lucro_total:,.2f}")
print(f"Margem de Lucro: {margem_lucro:.2f}%")
print(f"Ticket Médio: R$ {ticket_medio:,.2f}")
```

## 5. Métricas derivadas

### Tarefa

## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
# Agrupamento por canal de venda
resumo_canal = df.groupby('canal_venda').agg(
    receita_total=('valor_venda', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('valor_venda', 'count')
)

# Criar métricas derivadas
resumo_canal['margem'] = resumo_canal['lucro_total'] / resumo_canal['receita_total']
resumo_canal['ticket_medio'] = resumo_canal['receita_total'] / resumo_canal['quantidade_pedidos']

# Ordenar por receita (opcional)
resumo_canal = resumo_canal.sort_values(by='receita_total', ascending=False)

print(resumo_canal)
```

### Insight obrigatório
Escreva 1 ou 2 frases interpretando sua tabela, no estilo executivo:
- não apenas “o Online foi maior”
- mas o que isso **significa para a decisão**


```python
interpretacao = f"""
O canal {canal_mais_receita} lidera em volume de receita, indicando forte capacidade de geração de vendas,
enquanto o canal {canal_mais_eficiente} apresenta maior eficiência operacional, sugerindo melhor conversão em lucro.
Estratégicamente, vale escalar o canal líder em receita com otimizações de custo, enquanto se investe no canal mais eficiente para maximizar rentabilidade.
"""

print(interpretacao)
```

## 8. Análise por categoria

### Pergunta de negócio
**Qual categoria é mais lucrativa e qual parece mais eficiente?**

### Tarefa
Crie uma tabela agregada por `categoria` com:
- Receita Total

### Depois responda
1. A categoria de maior receita também é a de maior margem?

```python
# Agrupar por categoria
resumo_categoria = df.groupby('categoria').agg(
    receita_total=('valor_venda', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('valor_venda', 'count')
)

# Criar métricas derivadas
resumo_categoria['margem'] = resumo_categoria['lucro_total'] / resumo_categoria['receita_total']
resumo_categoria['ticket_medio'] = resumo_categoria['receita_total'] / resumo_categoria['quantidade_pedidos']

# Ordenar por receita
resumo_categoria = resumo_categoria.sort_values(by='receita_total', ascending=False)

print(resumo_categoria)
```

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total

- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?
2. Existe concentração regional?
3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?


```python
# Agrupar por UF
resumo_uf = df.groupby('uf').agg(
    receita_total=('valor_venda', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('valor_venda', 'count')
)

# Calcular participação percentual da receita
total_receita = resumo_uf['receita_total'].sum()
resumo_uf['participacao_%'] = (resumo_uf['receita_total'] / total_receita) * 100

# Ticket médio
resumo_uf['ticket_medio'] = resumo_uf['receita_total'] / resumo_uf['quantidade_pedidos']

# Ordenar por receita
resumo_uf = resumo_uf.sort_values(by='receita_total', ascending=False)

print(resumo_uf)
```

## 11. Campeões e vilões (Top N)

## 12. Análise temporal

A aula reforça que sem tempo a análise fica estática. fileciteturn5file0

### Tarefa
1. Converta `data_venda` para data, se necessário
2. Crie colunas auxiliares como:
   - mês
   - trimestre (opcional)
3. Gere uma tabela agregada por mês com:
   - receita
   - lucro
   - ticket médio ou margem

### Depois responda
1. O lucro parece crescer ao longo do tempo?
2. O último trimestre foi melhor que o anterior?
3. Quais meses concentram as vendas de fim de ano?


```python
# Garantir que a data está no formato correto
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# Criar colunas auxiliares
df['mes'] = df['data_venda'].dt.to_period('M')
df['trimestre'] = df['data_venda'].dt.to_period('Q')

# Tabela agregada por mês
resumo_mes = df.groupby('mes').agg(
    receita=('valor_venda', 'sum'),
    lucro=('lucro', 'sum'),
    pedidos=('valor_venda', 'count')
)

# Métricas derivadas
resumo_mes['ticket_medio'] = resumo_mes['receita'] / resumo_mes['pedidos']
resumo_mes['margem'] = resumo_mes['lucro'] / resumo_mes['receita']

print(resumo_mes)
```
