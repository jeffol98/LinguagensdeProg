```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

# Converter data
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# Criar colunas auxiliares para análise temporal
df['ano'] = df['data_venda'].dt.year
df['mes'] = df['data_venda'].dt.month
df['mes_nome'] = df['data_venda'].dt.strftime('%b')

# =====================================
# 1. GRÁFICO INTERATIVO - VENDAS POR CATEGORIA
# =====================================
fig_categoria = px.bar(
    df,
    x='categoria',
    y='receita',
    color='categoria',
    title='💰 Vendas por Categoria',
    hover_data=['produto', 'uf'], # Changed 'estado' to 'uf' as 'uf' is in df.columns
)

fig_categoria.show()

# =====================================
# 2. EVOLUÇÃO TEMPORAL (COM ZOOM INTERATIVO)
# =====================================
vendas_tempo = df.groupby('data_venda')['receita'].sum().reset_index()

fig_tempo = px.line(
    vendas_tempo,
    x='data_venda',
    y='receita',
    title='📈 Evolução das Vendas ao Longo do Tempo',
    markers=True
)

fig_tempo.update_layout(
    xaxis_title='Data',
    yaxis_title='Receita',
    hovermode='x unified'
)

fig_tempo.show()

# =====================================
# 3. RELAÇÃO ENTRE VARIÁVEIS (SCATTER INTERATIVO)
# =====================================
fig_relacao = px.scatter(
    df,
    x='quantidade',
    y='receita',
    color='categoria',
    size='receita',
    hover_data=['produto', 'uf'], # Changed 'estado' to 'uf' as 'uf' is in df.columns
    title='🔍 Relação entre Quantidade e Valor de Venda'
)

fig_relacao.show()

# =====================================
# 4. DISTRIBUIÇÃO ESPACIAL (MAPA DO BRASIL)
# =====================================
vendas_estado = df.groupby('uf')['receita'].sum().reset_index()

fig_mapa = px.choropleth(
    vendas_estado,
    locations='uf',
    locationmode='geojson-id',  # pode variar dependendo do dataset
    color='receita',
    scope='south america',
    title='🌎 Distribuição de Vendas por Estado'
)

fig_mapa.show()

# =====================================
# 5. DASHBOARD SIMPLES COM FILTRO (INTERATIVO)
# =====================================
# Exemplo: filtrar por categoria (simulando pergunta do gestor)

categoria_escolhida = df['categoria'].unique()[0]  # pode trocar manualmente

df_filtrado = df[df['categoria'] == categoria_escolhida]

fig_dashboard = px.line(
    df_filtrado.groupby('data_venda')['receita'].sum().reset_index(),
    x='data_venda',
    y='receita',
    title=f'📊 Vendas ao longo do tempo - {categoria_escolhida}'
)

fig_dashboard.show()
```

Por que esta atividade faz mais sentido em Plotly do que em Matplotlib?

Porque o Plotly é interativo, permitindo:

explorar os dados com zoom, hover e filtros
visualizar detalhes sem precisar criar vários gráficos separados
responder subperguntas na hora (ex: “qual produto vendeu mais nesse ponto?”)

Já o Matplotlib é estático, ou seja, você precisa gerar vários gráficos diferentes para cada nova análise.

Como essa atividade envolve exploração de dados e tomada de decisão, o Plotly é mais adequado.

Em que situação você ainda preferiria um gráfico estático?

Gráficos estáticos (como no Matplotlib) são melhores quando:

o objetivo é relatório formal (ex: artigo, TCC, PDF)
você precisa de padronização visual rigorosa
não há necessidade de interação
o gráfico será impresso

Exemplo: um relatório acadêmico ou documento oficial.

O destino desta análise parece mais “PDF” ou mais “tela/web”?

Claramente mais tela/web, porque:

há necessidade de interação
o gestor pode explorar os dados dinamicamente
os gráficos podem ser integrados em um dashboard

Ou seja, é um cenário típico de uso do Plotly.

## 4. Missão 1 — Barras: Receita por Canal

A missão prática da aula pede construir um gráfico de barras para responder:
**qual canal vende mais?**  
Os slides também sugerem usar o hover para injetar métricas secundárias sem poluir a tela. fileciteturn7file0

### Tarefa
1. Agregue a `receita` por `canal_venda`
2. Inclua também uma métrica secundária, como `quantidade`
3. Construa um gráfico de barras com Plotly Express
4. Ordene do maior para o menor
5. Faça o hover mostrar mais do que apenas a receita

### Perguntas
- Qual canal lidera a receita?
- O canal líder também lidera em quantidade?
- O hover ajudou a enriquecer a leitura sem poluir a tela?


```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

# Garantir formato de data (boa prática)
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# =====================================
# AGREGAÇÃO POR CANAL DE VENDA
# =====================================
df_canal = (
    df.groupby('canal_venda')
      .agg({
          'receita': 'sum',     # Receita total
          'quantidade': 'sum'       # Métrica secundária
      })
      .reset_index()
)

# =====================================
# ORDENAÇÃO (MAIOR → MENOR)
# =====================================
df_canal = df_canal.sort_values(by='receita', ascending=False)

# =====================================
# GRÁFICO DE BARRAS INTERATIVO
# =====================================
fig = px.bar(
    df_canal,
    x='canal_venda',
    y='receita',
    color='canal_venda',
    title='💰 Receita por Canal de Venda',

    # Hover enriquecido (métrica secundária + formatação)
    hover_data={
        'receita': ':.2f',
        'quantidade': True
    },

    labels={
        'canal_venda': 'Canal de Venda',
        'receita': 'Receita Total',
        'quantidade': 'Quantidade Vendida'
    }
)

# Melhorar aparência e legibilidade
fig.update_layout(
    xaxis_title='Canal de Venda',
    yaxis_title='Receita Total',
    showlegend=False
)

# =====================================
# EXIBIR GRÁFICO
# =====================================
fig.show()
```

Qual canal lidera a receita?

O canal que aparece no topo do gráfico (barra mais alta após a ordenação) é o que lidera a receita total.
Como o gráfico já está ordenado do maior para o menor, basta olhar o primeiro colocado no eixo X.

O canal líder também lidera em quantidade?

Nem sempre.
Ao analisar o hover, é possível comparar a quantidade vendida com a receita:

Se o mesmo canal tiver também a maior quantidade → ele lidera em ambos
Caso contrário → indica diferença de estratégia, como:
preços mais altos (mais receita, menos quantidade)
maior volume com ticket médio menor
O hover ajudou a enriquecer a leitura sem poluir a tela?

Sim, totalmente.
O hover permite:

visualizar informações adicionais (quantidade) sob demanda
manter o gráfico limpo e focado na métrica principal (receita)
facilitar a exploração sem precisar criar vários gráficos

Ou seja, ele melhora a análise sem comprometer a clareza visual.

## 5. Missão 2 — Linhas: Sazonalidade da Receita Mensal

Os slides destacam que o gráfico de linhas, com zoom e pan, é ideal para navegar no tempo e isolar picos. Também reforçam que o eixo temporal precisa estar corretamente tipado no Pandas. fileciteturn7file0

### Tarefa
1. Converta `data_venda` em data, se necessário
2. Agregue a `receita` por mês
3. Construa um gráfico de linha interativo

### Perguntas

- O zoom ajudou a explorar melhor a parte final da série?


```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

# Garantir que data_venda é datetime
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# Remover possíveis datas inválidas
df = df.dropna(subset=['data_venda'])

# =====================================
# AGREGAÇÃO POR MÊS
# =====================================
# Criar coluna ano-mês (período)
df['ano_mes'] = df['data_venda'].dt.to_period('M').astype(str)

# Agrupar receita por mês
df_mes = (
    df.groupby('ano_mes')['receita']
      .sum()
      .reset_index()
)

# Converter de volta para datetime (melhora eixo do Plotly)
df_mes['ano_mes'] = pd.to_datetime(df_mes['ano_mes'])

# Ordenar corretamente
df_mes = df_mes.sort_values(by='ano_mes')

# =====================================
# GRÁFICO DE LINHA INTERATIVO
# =====================================
fig = px.line(
    df_mes,
    x='ano_mes',
    y='receita',
    title='📈 Receita Mensal ao Longo do Tempo',
    markers=True
)

# Melhorias de interatividade
fig.update_layout(
    xaxis_title='Mês',
    yaxis_title='Receita',
    hovermode='x unified'
)

fig.show()
```

Existe sazonalidade?

Sim, há indícios de sazonalidade se o gráfico mostra aumentos recorrentes em determinados períodos, especialmente no final do ano. Isso sugere que as vendas não são distribuídas de forma uniforme ao longo dos meses.

Quais meses chamam atenção?

Os meses que apresentam picos mais elevados — geralmente novembro e dezembro — chamam mais atenção, possivelmente devido a eventos como promoções e datas comemorativas.

O zoom ajudou a explorar melhor a parte final da série?

Sim. O zoom permite focar nos meses finais, facilitando a identificação de picos e comparações mais detalhadas, algo que seria difícil em um gráfico estático.

```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

# Converter data (boa prática)
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# =====================================
# CRIAR MÉTRICA DE MARGEM (SE NECESSÁRIO)
# =====================================
# margem_lucro = lucro / receita
if 'margem_lucro' not in df.columns:
    df['margem_lucro'] = df['lucro'] / df['receita'] # Corrected: valor_venda to receita

# =====================================
# SCATTER PLOT INTERATIVO
# =====================================
fig = px.scatter(
    df,
    x='receita',          # Corrected: valor_venda to receita
    y='lucro',                # Lucro
    color='categoria',        # Cor por categoria
    title='🔍 Receita vs Lucro por Categoria',

    # Hover enriquecido
    hover_data={
        'produto': True,
        'canal_venda': True,
        'uf': True,
        'margem_lucro': ':.2f',
        'receita': ':.2f', # Corrected: valor_venda to receita
        'lucro': ':.2f'
    }
)

# Melhorias visuais
fig.update_layout(
    xaxis_title='Receita',
    yaxis_title='Lucro',
    hovermode='closest'
)

fig.show()
```

## 8. A tríade da interatividade

Um slide central da aula apresenta três componentes principais:
- **Tooltips (hover)**
- **Zoom & Pan**
- **Filtros visuais / isolamento via legenda** fileciteturn7file0

### Tarefa
Escolha um dos seus gráficos interativos e descreva, em markdown:
1. O que o hover acrescenta
2. Como o zoom melhora a investigação
3. Como o clique na legenda pode ajudar a isolar séries ou categorias


```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# =====================================
# GRÁFICO INTERATIVO COMPLETO
# (TOOLTIPS + ZOOM + FILTRO POR LEGENDA)
# =====================================
fig = px.scatter(
    df,
    x='receita',
    y='lucro',
    color='categoria',  # legenda interativa (filtro)
    size='quantidade',

    title='🎯 Receita vs Lucro (Interativo Completo)',

    # TOOLTIP (HOVER RICO)
    hover_data={
        'produto': True,
        'canal_venda': True,
        'uf': True,
        'margem_lucro': ':.2f',
        'receita': ':.2f',
        'lucro': ':.2f'
    }
)

# =====================================
# MELHORIAS DE INTERAÇÃO
# =====================================
fig.update_layout(
    xaxis_title='Receita',
    yaxis_title='Lucro',
    hovermode='closest'
)

# =====================================
# EXIBIR
# =====================================
fig.show()
```

O hover acrescenta contexto detalhado sem poluir o gráfico, permitindo ver informações como produto, canal, UF e margem para cada ponto.

O zoom melhora a investigação ao possibilitar focar em regiões específicas do gráfico, facilitando a análise de padrões, clusters e possíveis outliers.

O clique na legenda permite ativar ou ocultar categorias, ajudando a isolar séries e comparar grupos de forma clara e dinâmica.

### 📊 O que significa dizer que Plotly Express é “declarativo”?

Significa que você descreve **o que quer visualizar**, e não como construir cada detalhe do gráfico.  
Ou seja, basta informar colunas (eixo X, Y, cor, tamanho, etc.) e o Plotly Express cuida automaticamente da estrutura e formatação.

---

### ⚡ Em que ele facilita a vida do analista?

Ele reduz muito a quantidade de código e decisões manuais, permitindo:
- criar gráficos complexos com poucas linhas  
- padronizar visualizações rapidamente  
- focar mais na **análise dos dados** do que na implementação  

---

### 🌐 Como isso se conecta com a ideia de futuro dashboard em Streamlit?

Como o Plotly Express já trabalha de forma declarativa, fica muito fácil integrar os gráficos em um app no Streamlit.  
Basta conectar os dados a **inputs interativos (filtros, seleções)** e o gráfico se atualiza automaticamente, viabilizando dashboards dinâmicos com pouco esforço.

```python
import pandas as pd
import plotly.express as px

# Leitura
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# =====================================
# AGREGAÇÃO
# =====================================
df_canal = (
    df.groupby('canal_venda')
      .agg({
          'receita': 'sum', # Changed 'valor_venda' to 'receita'
          'quantidade': 'sum'
      })
      .reset_index()
      .sort_values(by='receita', ascending=False) # Changed 'valor_venda' to 'receita'
)

# =====================================
# GRÁFICO MELHORADO
# =====================================
fig = px.bar(
    df_canal,
    x='canal_venda',
    y='receita', # Changed 'valor_venda' to 'receita'
    color='canal_venda',

    # TÍTULO AUTOEXPLICATIVO
    title='💰 Receita Total por Canal de Venda (R$)',

    # HOVER MELHORADO
    hover_data={
        'canal_venda': False,  # evita redundância
        'receita': ':.2f', # Changed 'valor_venda' to 'receita'
        'quantidade': True
    },

    # NOMES AMIGÁVEIS
    labels={
        'canal_venda': 'Canal de Venda',
        'receita': 'Receita Total (R$)', # Changed 'valor_venda' to 'receita'
        'quantidade': 'Quantidade Vendida'
    }
)

# =====================================
# AJUSTES VISUAIS (LIMPEZA)
# =====================================
fig.update_layout(
    showlegend=False,  # remove legenda redundante
    xaxis_title='Canal de Venda',
    yaxis_title='Receita Total (R$)',
    template='plotly_white'  # fundo limpo
)

# Ajustar ordem das barras
fig.update_xaxes(categoryorder='total descending')

fig.show()
```

```python
# =====================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================
import pandas as pd
import plotly.express as px

# =====================================
# LEITURA E TRATAMENTO DOS DADOS
# =====================================
df = pd.read_csv("vendas_brasil_clean_aula6_plotly.csv")

# Converter datas
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# Remover valores inválidos de data
df = df.dropna(subset=['data_venda'])

# Criar coluna de margem (se não existir)
if 'margem_lucro' not in df.columns:
    df['margem_lucro'] = df['lucro'] / df['receita'] # Corrected: valor_venda to receita

# =====================================
# 1. BARRAS — RECEITA POR CANAL
# =====================================
df_canal = (
    df.groupby('canal_venda')
      .agg({'receita': 'sum', 'quantidade': 'sum'}) # Corrected: valor_venda to receita
      .reset_index()
      .sort_values(by='receita', ascending=False) # Corrected: valor_venda to receita
)

fig_bar = px.bar(
    df_canal,
    x='canal_venda',
    y='receita', # Corrected: valor_venda to receita
    color='canal_venda',
    title='💰 Receita Total por Canal de Venda (R$)',
    hover_data={'receita': ':.2f', 'quantidade': True}, # Corrected: valor_venda to receita
    labels={
        'canal_venda': 'Canal',
        'receita': 'Receita (R$)', # Corrected: valor_venda to receita
        'quantidade': 'Quantidade'
    }
)

fig_bar.update_layout(
    showlegend=False,
    template='plotly_white'
)

fig_bar.show()

# =====================================
# 2. LINHAS — SAZONALIDADE
# =====================================
df['ano_mes'] = df['data_venda'].dt.to_period('M').astype(str)

df_mes = (
    df.groupby('ano_mes')['receita'] # Corrected: valor_venda to receita
      .sum()
      .reset_index()
)

df_mes['ano_mes'] = pd.to_datetime(df_mes['ano_mes'])
df_mes = df_mes.sort_values(by='ano_mes')

fig_line = px.line(
    df_mes,
    x='ano_mes',
    y='receita', # Corrected: valor_venda to receita
    title='📈 Evolução Mensal da Receita',
    markers=True
)

fig_line.update_layout(
    hovermode='x unified',
    template='plotly_white'
)

fig_line.show()

# =====================================
# 3. DISPERSÃO — RECEITA vs LUCRO
# =====================================
fig_scatter = px.scatter(
    df,
    x='receita', # Corrected: valor_venda to receita
    y='lucro',
    color='categoria',
    size='receita', # Corrected: valor_venda to receita
    title='🔍 Receita vs Lucro por Categoria',
    hover_data={
        'produto': True,
        'canal_venda': True,
        'uf': True,
        'margem_lucro': ':.2f',
        'receita': ':.2f' # Adding receita to hover_data for consistency
    }
)

fig_scatter.update_layout(
    template='plotly_white'
)

fig_scatter.show()

# =====================================
# 4. MAPA — DISTRIBUIÇÃO GEOGRÁFICA
# =====================================
fig_map = px.scatter_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    size='receita', # Corrected: valor_venda to receita
    color='receita', # Corrected: valor_venda to receita
    hover_data={
        'uf': True, # Corrected 'estado' and removed 'cidade'
        'receita': ':.2f' # Corrected: valor_venda to receita
    },
    zoom=3,
    title='🌎 Distribuição Geográfica das Vendas'
)

fig_map.update_layout(
    mapbox_style='open-street-map',
    template='plotly_white'
)

fig_map.show()
```
