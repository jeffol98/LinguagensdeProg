### Tarefa

```python
df['uf'].unique()
df['canal'].unique()
df['categoria'].unique()

# Remover espaços extras e padronizar caixa
df['uf'] = df['uf'].str.strip().str.upper()

df['canal'] = df['canal'].str.strip().str.lower()
df['categoria'] = df['categoria'].str.strip().str.lower()
```

## 8. Subindo de nível — Criação de variáveis derivadas

Depois da limpeza, é hora de enriquecer a base com variáveis úteis para análise. fileciteturn2file0

### Tarefa
Crie, se possível:
- `ano`
- `mes`
- `ano_mes`
- `margem_lucro`

### Atenção
A criação de `margem_lucro` exige cuidado com divisões por zero e possíveis valores infinitos.


```python
# Garantindo que a coluna já está em datetime
df['ano'] = df['data'].dt.year
df['mes'] = df['data'].dt.month

# Formato ano-mês (bom para agrupamentos)
df['ano_mes'] = df['data'].dt.to_period('M').astype(str)

# Criando marquem de lucro
# Evitar divisão por zero
df['margem_lucro'] = df['lucro'] / df['receita']

# Tratar infinitos e valores inválidos
df['margem_lucro'] = df['margem_lucro'].replace([np.inf, -np.inf], np.nan)

```

### Reflexão técnica
Explique:
1. O que pode acontecer ao calcular margem de lucro quando a receita é zero?
2. Como você decidiu tratar esse caso?


Problema: margem de lucro com receita zero

A margem de lucro é calculada como:

lucro / receita
⚠️ O que acontece quando a receita é zero?
❌ Ocorre divisão por zero
Isso pode gerar:
inf (infinito)
-inf
ou erro dependendo do contexto
🚨 Impacto na análise:
Distorce médias e indicadores
Pode “quebrar” gráficos
Prejudica KPIs como margem média
Pode passar despercebido e gerar conclusões erradas
🧠 Decisão de tratamento
✔️ Estratégia adotada:
Quando receita = 0, a margem de lucro foi definida como NaN (nulo)
df['margem_lucro'] = np.where(
    df['receita'] > 0,
    df['lucro'] / df['receita'],
    np.nan
)
🧾 Justificativa

Quando a receita é zero, não é possível calcular a margem de lucro de forma válida, pois não há base de comparação para o lucro. Manter valores infinitos poderia distorcer análises e indicadores.
Por isso, optou-se por substituir esses casos por valores nulos (NaN), garantindo maior consistência e confiabilidade nos resultados.
